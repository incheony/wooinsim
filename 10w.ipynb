```python
def selection_sort(A):
    n = len(A)
    for i in range(n-1):
        least = i;
        for j in range(i+1, n):
            if (A[j]<A[least]):
                least = j
        A[i], A[least] = A[least], A[i]
        print("Step", i+1, ":", A)
```

```python
def sortGapInsertion(A, first, last, gap):
    for i in range(first+gap, last+1, gap):
        key = A[i]
        j = i-gap
        while j>=first and key<A[j]:
            A[j+gap] = A[j]
            j = j-gap
        A[j+gap]= key

def shell_sort(A):
    n = len(A)
    gap = n//2
    while gap > 0:
        if (gap % 2) == 0:
            gap += 1
        for i in range(gap):
            sortGapInsertion(A, i, n-1, gap)
        print('Gap=', gap, ":",A)
        gap = gap//2
```

```python
# Selection Sort

print("마라토너의 수를 입력하시오 : ", end='')
player = int(input())

print("마라토너의 기록을 초단위로 입력 : ")
records = [0]*player
for i in range(player):
    records[i] = int(input())
    
print("마라토너의 선수 번호 입력 : ")
numbers = [0]*player
for n in range(player):
    numbers[n] = int(input())
    
records_numbers = dict(zip(records, numbers))   

print('')    
#print("**(Selection_sort)**")
print("정렬 전:", records)
selection_sort(records)
print("정렬 후:", records)

print('')
for k in range(3):
    hour = records[k] // (60*60)
    min = (records[k]-hour*60*60) // 60
    sec = records[k]%60
    num = records_numbers[records[k]]
    print("%1d 등 : 선수 번호(%d) %1d시간 %2d분 %2d초"%(k+1,num, hour, min, sec))
```

```python
# Shell Sort

print("마라토너의 수를 입력하시오 : ", end='')
player = int(input())

print("마라토너의 기록을 초단위로 입력 : ")
records = [0]*player
for i in range(player):
    records[i] = int(input())
    
print("마라토너의 선수 번호 입력 : ")
numbers = [0]*player
for n in range(player):
    numbers[n] = int(input())
    
records_numbers = dict(zip(records, numbers))   

print('')    
print("**(Shell_sort)**")
print("정렬 전:", records)
shell_sort(records)
print("정렬 후:", records)

print('')
for k in range(3):
    hour = records[k] // (60*60)
    min = (records[k]-hour*60*60) // 60
    sec = records[k]%60
    num = records_numbers[records[k]]
    print("%1d 등 : 선수 번호(%d) %1d시간 %2d분 %2d초"%(k+1,num, hour, min, sec))
```
